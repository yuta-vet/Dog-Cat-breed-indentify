下の方にファイルを選択するボタンが出てきますので、それを押して解析したい写真を選択してください。

```python
from google.colab import files;import numpy as np;import tensorflow as tf;import matplotlib.pyplot as plt;import matplotlib.image as mpimg;from tensorflow.keras.preprocessing.image import load_img, img_to_array
uploaded = files.upload()
plt.axis('off');plt.imshow(mpimg.imread("/content/" + list(uploaded.keys())[0]));plt.show()
processedpic = tf.keras.applications.efficientnet.preprocess_input(np.expand_dims(img_to_array(load_img("/content/" + list(uploaded.keys())[0], target_size= (224,224,3))).astype('float32'),axis=0))
prediction = tf.keras.applications.efficientnet.decode_predictions(tf.keras.applications.EfficientNetB0(weights='imagenet').predict(processedpic))
for breed in [item[1] for rank,item in enumerate(prediction[0]) if rank <= 0 ]: print("\n予測される種類は",breed,"です。")
```
